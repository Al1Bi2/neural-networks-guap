```python
from sklearn.model_selection import train_test_split
def split_data(data, column_number):
    data_y = data[column_number]
    data_x = data.drop(columns=[column_number])
    test_size = 0.2
    val_size = 0.25
    seed = 7
    
    x_train, x_test, y_train, y_test = train_test_split(data_x,data_y,test_size = test_size, random_state = seed)
    
    x_train2, x_val, y_train2, y_val = train_test_split(x_train,y_train,test_size = val_size, random_state = seed)
    
    return x_train,x_test,x_val, y_train,y_test, y_val
```

Создадим фунцию для разделения набора данных

```python
full_x_train,full_x_test, full_x_val,full_y_train,full_y_test, full_y_val=split_data(concrete_full,"concrete_compressive_strength")
print(full_x_train.shape[0],full_x_val.shape[0],full_x_test.shape[0])
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten, Input

def build_fully_connected_model(input_shape):
    model = Sequential()
    model.add(Input(shape=input_shape))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

def build_cnn_model(input_shape):
    model = Sequential()
    model.add(Input(shape=input_shape))
    model.add(Conv1D(64, kernel_size=2, activation='relu'))
    model.add(Flatten())
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1)) 
    model.compile(optimizer='adam', loss='mse')
    return model

def evaluate_model(model, x_train, y_train, x_val, y_val):
    model.fit(x_train, y_train, epochs=100, batch_size=32, verbose=0)

    y_train_pred = model.predict(x_train)
    y_val_pred = model.predict(x_val)

    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))

    train_r2 = r2_score(y_train, y_train_pred)
    val_r2 = r2_score(y_val, y_val_pred)

    return train_rmse, val_rmse, train_r2, val_r2
```

```python
# Полносвязная модель
fc_model = build_fully_connected_model((full_x_train.shape[1],))
fc_model.fit(full_x_train, full_y_train, epochs=100, batch_size=32, validation_data=(full_x_val, full_y_val))

fc_predictions = fc_model.predict(full_x_val)
fc_rmse = np.sqrt(mean_squared_error(full_y_val, fc_predictions))
fc_r2 = r2_score(full_y_val, fc_predictions)

print(f'Полносвязная модель - RMSE: {fc_rmse}, R2: {fc_r2}')
```

```python
# Сверточная модель
full_x_train_cnn = full_x_train.to_numpy().reshape((full_x_train.shape[0], full_x_train.shape[1], 1))
full_x_val_cnn = full_x_val.to_numpy().reshape((full_x_val.shape[0], full_x_val.shape[1], 1))

cnn_model = build_cnn_model((full_x_train_cnn.shape[1], 1))
cnn_model.fit(full_x_train_cnn, full_y_train, epochs=100, batch_size=32, validation_data=(full_x_val_cnn, full_y_val))

cnn_predictions = cnn_model.predict(full_x_val_cnn)
cnn_rmse = np.sqrt(mean_squared_error(full_y_val, cnn_predictions))
cnn_r2 = r2_score(full_y_val, cnn_predictions)

print(f'Сверточная модель - RMSE: {cnn_rmse}, R2: {cnn_r2}')
```
